```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame as df
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
data
```

```python
data.head()
```

```python
data.describe()
```

```python
# separating data int cost and revenue

X = df(data, columns = ['production_budget_usd']) # feature or independent variable
y = df(data, columns = ['worldwide_gross_usd']) # depentent variable or the target
```

```python
# Visualizing the data using 'matplotlib'

plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)

plt.title('Film cost vs Global rev.')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)

plt.show()
```

```python
# Applying the linear regression for 

regression = LinearRegression()
regression.fit(X, y)
```

```python
# Slope co-efficient
# y = mX + c

regression.coef_  # m
```

```python
regression.intercept_ # c
```

```python
# Visualizing chart with the regression line

plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.plot(X, regression.predict(X), color='red', linewidth=4)

plt.title('Film cost vs Global rev.')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)

plt.show()
```

```python
regression.score(X, y)
```
